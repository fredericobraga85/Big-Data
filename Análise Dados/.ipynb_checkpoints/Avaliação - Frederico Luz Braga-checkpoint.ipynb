```python
#libs necessários
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import time
import datetime
```

```python
#Buscar data de arquivo
data = train = pd.read_csv('PRSA_data_2010.1.1-2014.12.31.csv') 
data.head()
```

### As colunas da tabela possuem os seguintes significados:

- No: número da linha
- year: ano 
- month: mês 
- day: dia
- hour: hora
- pm2.5: concetração PM2.5 (ug/m^3) 
- DEWP: Dew Point (â„ƒ)*
- TEMP: Temperatura(â„ƒ) 
- PRES: Pressão (hPa) 
- cbwd: Direção do vento
- Iws: Velocidade do vento acumulada(m/s) 
- Is: Horas de nevasca acumuladas
- Ir: Horas de chuva acumuladas


* O DEW Point é conhecido como o 'Ponto de orvalho', que  designa a temperatura a qual o vapor de água presente no ar ambiente passa ao estado líquido na forma de pequenas gotas por via da condensação, o chamado orvalho (Ref: https://pt.wikipedia.org/wiki/Ponto_de_orvalho)

```python
#métodos utils

def convertToTimestmap(d,m,y, h):
    
    date = str(d) + "/" + str(m) + "/" + str(y) + " " + str(h) + ":00" 
    return time.mktime(datetime.datetime.strptime(date, "%d/%m/%Y %H:00" ).timetuple())
```

### 1) Apresente os dados, concentração e dispersão quanto á chuva e nevascas na região de Hong Kong:

```python
#mostrar acumulado de horas de chuva por mês

print data.groupby(['month'])['Ir'].sum()

plt.plot(data.groupby(['month'])['Ir'].sum())
plt.title('Horas de chuva por mes (entre 2010 e 2014)')
plt.show()
```

É possível observar que Julho (1887 horas) ,Agosto (953 horas) Setembro (1859 horas) são os meses com mais chuva. E que Dexembro ( 7 horas) , Janeiro (15 horas) e Fevereiro(7 horas) são os meses com menos horas de chuva.

```python
#mostrar acumulado de horas de neve por mês

print data.groupby(['month'])['Is'].sum()

plt.plot(data.groupby(['month'])['Is'].sum())
plt.title('Horas de neve por mes (entre 2010 e 2014)')
plt.show()
```

É possível observar que neva em Hong Kong somente em Novembro (4 horas) , Dezembro (376 horas) , Janeiro (760 horas) , Fevereiro ( 840 horas) e Março (331 horas)

### 2) Apresente os dados, concentração e dispersão quanto ao volume de partículas suspensas (PM2.5) na atmosfera de Hong Kong. 

```python
#mostrar média de pm2.5 por dia

range = 16

df_avg_pm25_day =  data.groupby(['year','month', 'day'])['pm2.5'].mean()

n, bins, patches = plt.hist(df_avg_pm25_day.dropna(0), bins=range , range=(0,df_avg_pm25_day.max()) , rwidth='scalar')
plt.axvline(35, color='r', linestyle='dashed', linewidth=2)

plt.title('Media de Concentracao de pm2.5 por dia (entre 2010 e 2014)')
plt.show()

print 'Média de PM2.5 por dia:', df_avg_pm25_day.dropna().mean()
print 'Desvio Padrao PM2.5 por dia:' , df_avg_pm25_day.dropna().std()

```

De acordo com US Environmental Protection Agency’s (EPA), o limite aceitável de concentração de PM2.5 durante 24 horas é de 35 ug/m3 (representado com a linha vermelha).

A média por dia em Hong Kong é de 98 ug/m3. Três vezes mais que o limite.

### 3) Demostre se há diferenças significativas na concentração de PM2.5 quanto à estação do ano, temperatura, chuva, neve e períodos (manhã, tarde e noite).

```python
# Análise de concentração de PM2.5 quanto à estação do ano\

inverno = [1,11,12]
primavera = [2,3,4]
verao = [5,6,7]
outono = [8,9,10]

df_inverno   = data.groupby(['month']).mean()[['pm2.5']].loc[inverno].mean()
df_primavera = data.groupby(['month']).mean()[['pm2.5']].loc[primavera].mean()
df_verao     = data.groupby(['month']).mean()[['pm2.5']].loc[verao].mean()
df_outono    = data.groupby(['month']).mean()[['pm2.5']].loc[outono].mean()

l = np.array([df_inverno.T.iloc[0].tolist(),
              df_primavera.T.iloc[0].tolist(),
              df_verao.T.iloc[0].tolist(),
              df_outono.T.iloc[0].tolist()])

df = pd.DataFrame(l)

df.plot.bar(legend=False, color='r');
ind = np.arange(4)
plt.xticks(ind, ('Inverno', 'Primavera', 'Verao', 'Outono'))
plt.title('Concenrtracao PM2.5 x Estacao do ano')
plt.show()
```

É possível observar que existe uma maior concetração de PM2.5 durante o inverno e uma menor durante o verão.

```python
# Análise de concentração de PM2.5 quanto à Temperatura \



df_temp   = data.groupby(['TEMP'])['pm2.5'].mean()

fig, ax1 = plt.subplots()

ax1.plot(df_temp, 'b-')
ax1.set_xlabel('temperatura')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('Media concentracao PM2.5')


plt.title('Concenrtracao PM2.5 x Temperatura')


plt.show()


```

É possível observar que existe uma maior concetração de PM2.5 nas temperaturas mais baixas. Em temperaturas de -3 graus , observa-se uma  média acima de 120ug/m3 , quatro vezes acima do limite estabelcido pelo US Environmental Protection Agency’s (EPA).

```python
# Análise de concentração de PM2.5 com a chuva a cada mês\

df_sum_chuva_month = data.groupby(['month'])['Ir'].sum()
df_avg_pm25_month  =  data.groupby(['month'])['pm2.5'].mean()

fig, ax1 = plt.subplots()

ax1.plot(df_sum_chuva_month, 'b-')
ax1.set_xlabel('mes')
# Make the y-axis label, ticks and tick labels match the line color.
ax1.set_ylabel('horas chuva', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot( df_avg_pm25_month, 'r--')
ax2.set_ylabel('concentracao pm2.5', color='r')
ax2.tick_params('y', colors='r')

plt.title('Horas de chuva x Concenrtracao PM2.5 x Mes (entre 2010 e 2014)')

fig.tight_layout()
plt.show()
```

É possível observar que existe uma maior concetração de PM2.5 em meses com menos chuvas
